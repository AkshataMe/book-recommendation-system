```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings
```

```python
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number of rating'},inplace=True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating
```

```python
final_rating = final_rating[final_rating['number of rating']>=50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivote = final_rating.pivot_table(columns='user_id',index='title',values='rating')
book_pivote
```

```python
book_pivote.fillna(0,inplace=True)
book_pivote
```

```python
from scipy.sparse import csr_matrix

book_sparse = csr_matrix(book_pivote)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestions = model.kneighbors(book_pivote.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivote.index==book_name)[0][0]
    distance, suggestions = model.kneighbors(book_pivote.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(len(suggestions)):
        if i == 0:
            print('the suggestion for', book_name, 'are : ')
        if not i:
              print(book_pivote.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
